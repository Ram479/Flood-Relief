```python
from tkinter import *
from tkinter import messagebox

root = Tk()

def getvals():
    def donated():
        with open("donate.txt","r") as d:
            dtd = d.read()
            messagebox.showinfo("Donated money", dtd)
        
    def relief():
        messagebox.showinfo("fyytft", "tftrfytftrtdestufi7tfrr")
        
    def donate():
            with open("donate.txt","a") as f:
                f.write("Name: "+nameentry.get() +"\n")
                f.write("Mobile: "+mobileentry.get()+"\n")
                f.write("Amount: "+amountentry.get()+"\n"+"\n")
            Label(ui,text="Money donated").place(x=250,y=320)
                
                
    if namevalue.get()=='admin' and passvalue.get()=='admin':
        ui=Toplevel()
        ui.geometry('800x600')
        a=Canvas(ui,height=600,width=800,bd=2)
        img2=PhotoImage(file="backgmain.png")
        a.create_image(0,0,anchor=NW,image=img2)
        a.place(x=0,y=0)

        
        name = Label(ui, text="Name",font=("corbel",20,"bold"))
        mobile = Label(ui, text="Mobile Number",font=("corbel",20,"bold"))
        amount = Label(ui, text="Amount",font=("corbel",20,"bold"))
        name.place(x=30,y=50)
        mobile.place(x=30,y=80)
        amount.place(x=30,y=110)
        
        namevalue1=StringVar()
        mobilevalue = StringVar()
        amountvalue = IntVar()
        
        nameentry = Entry(ui, textvariable=namevalue1,font=("corbel",20,"bold"))
        mobileentry = Entry(ui, textvariable=mobilevalue,font=("corbel",20,"bold"))
        amountentry = Entry(ui, textvariable=amountvalue,font=("corbel",20,"bold"))
        nameentry.place(x=350,y=50)
        mobileentry.place(x=350,y=80)
        amountentry.place(x=350,y=110)
        
        Button(ui,text="Submit", command=donate,font=("corbel",20,"bold")).place(x=560, y=170)
        Button(ui,text="Relief Center", command=relief,font=("corbel",20,"bold")).place(x=50, y=500)
        Button(ui,text="Donated money", command=donated,font=("corbel",20,"bold")).place(x=480, y=500)
        ui.mainloop()

        
        
    else:
        messagebox.showinfo("Error","Password incorrect \nPlease enter again")


root.geometry("500x300")
root.maxsize(500,300)
root.title("Flood Relief System")
c=Canvas(root,height=300,width=500,bg="black",bd=0)
img=PhotoImage(file="backg.png")
c.create_image(0,0,anchor=NW,image=img)
c.place(x=0,y=0)

#Text for our form
name = Label(root, text="Username",font=("corbel",20,"bold"))
password = Label(root, text="Password",font=("corbel",20,"bold"))


#Pack text for our form
name.place(x=20,y=80)
password.place(x=20,y=130)


# Tkinter variable for storing entries
namevalue = StringVar()
passvalue = StringVar()

#Entries for our form
nameentry = Entry(root, textvariable=namevalue,font=("corbel",20,"bold"))
passentry = Entry(root, textvariable=passvalue,show="*",font=("corbel",20,"bold"))

# Packing the Entries
nameentry.place(x=170,y=80)
passentry.place(x=170,y=130)
#Button & packing it and assigning it a command
Button(text="Submit", command=getvals,font=("corbel",20,"bold")).place(x=340,y=210)
root.mainloop()
```
